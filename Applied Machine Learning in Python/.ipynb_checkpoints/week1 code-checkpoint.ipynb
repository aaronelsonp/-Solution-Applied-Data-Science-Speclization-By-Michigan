```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
fruits=pd.read_table(r"C:\Users\MMOHTASHIM\Anaconda3\libs\Applied Data Science Speclization-Michigan\Applied Machine Learning in Python\course3_downloads\fruit_data_with_colors.txt")
```

```python
fruits.head()
```

```python
X=fruits[["mass","width","height","color_score"]]
y=fruits["fruit_label"]
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
from matplotlib import cm
cmap=cm.get_cmap('gnuplot')
scatter=pd.scatter_matrix(X_train,c=y_train,marker='o',s=40,hist_kwds={"bins":15},figsize=(12,12),cmap=cmap)
```

```python
from mpl_toolkits.mplot3d import axes3d
fig=plt.figure()
ax=fig.add_subplot(111,projection="3d")
ax.scatter(X_train["width"],X_train["height"],X_train["color_score"],c=y_train,marker="o",s=100)
ax.set_xlabel("width")
ax.set_ylabel("height")
ax.set_zlabel("color_score")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
lookup_fruit=dict(zip(fruits.fruit_label.unique(),fruits.fruit_name.unique()))
fruit_prediction=knn.predict([[20,4.3,5.5,0.53]])
lookup_fruit[fruit_prediction[0]]

```
