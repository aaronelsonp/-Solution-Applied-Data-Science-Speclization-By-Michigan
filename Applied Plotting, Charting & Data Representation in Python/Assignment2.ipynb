```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd
%matplotlib inline
def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(18,'172a2fdce873f964749851f7565d3b3b189fad6bce734e5a5eaf082d')
```

```python
import datetime as dt
df.dtypes
time_format = '%Y-%m-%d'

df['Date'] = pd.to_datetime(df['Date'], format=time_format)  

df.set_index('Date',inplace=True)
df= df.sort_index()
df.reset_index(inplace=True)
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df
```

```python
df_max=df[(df["Element"]=="TMAX")&(df["year"]!=2015)]
df_min=df[(df["Element"]=="TMIN")&(df["year"]!=2015)]
df_max_2015=df[(df["Element"]=="TMAX")&(df["year"]==2015)]
df_min_2015=df[(df["Element"]=="TMIN")&(df["year"]==2015)]
```

```python
import numpy as np 
temp_min=df_min.groupby(df_min['Date'].dt.strftime('%m-%d')).agg({'Data_Value': np.min})
temp_max=df_max.groupby(df_max['Date'].dt.strftime('%m-%d')).agg({'Data_Value': np.max})
temp_min_2015=df_max_2015.groupby(df_max_2015['Date'].dt.strftime('%m-%d')).agg({'Data_Value': np.min})
temp_max_2015=df_min_2015.groupby(df_min_2015['Date'].dt.strftime('%m-%d')).agg({'Data_Value': np.max})
```

```python
temp_max_2015["Data_Value_2"]=temp_max["Data_Value"]
temp_min_2015["Data_Value_2"]=temp_min["Data_Value"]
min_2015=temp_min_2015[(temp_min_2015["Data_Value"]<temp_min_2015["Data_Value_2"])]["Data_Value"].tolist()
min_2015_index=temp_min_2015[(temp_min_2015["Data_Value"]<temp_min_2015["Data_Value_2"])].index.tolist()
max_2015=temp_max_2015[(temp_max_2015["Data_Value"]>temp_max_2015["Data_Value_2"])]["Data_Value"].tolist()
max_2015_index=temp_max_2015[(temp_max_2015["Data_Value"]>temp_max_2015["Data_Value_2"])].index.tolist()

(min_2015,min_2015_index)
```

```python
plt.style.use('seaborn-notebook')
ax=plt.gca()
fig=plt.gcf()
ax.plot(temp_min["Data_Value"].tolist(),color='g',label='TempMin(2005-2014)')
ax.plot(temp_max["Data_Value"].tolist(),color='r',label='TempMax(2005-2014)')
ax.set(xticks=np.arange(10,370,30),xticklabels=["Jan", "Feb", "Mar","Apr","May","June",'July','Aug',
                                                      'Sep','Oct','Nov','Dec'])
ax.grid(False)
ax.set_axis_bgcolor('white')
plt.fill_between(np.arange(0,366),np.array(temp_min["Data_Value"]),np.array(temp_max["Data_Value"]),color='grey',alpha=0.4)
plt.xlabel('Months',fontsize=15)
plt.ylabel('Value',fontsize=15)
plt.title("Max and Min Temp of Lahore,Pakistan between 2005-2014 with 2015 outlier values",fontsize=20)
plt.scatter([120,135,136,247,351],max_2015,color='r',alpha=0.8,marker="^",s=200,label='2015 Max record')
plt.scatter([55,120],min_2015,color='g',alpha=0.8,marker="v",s=200,label='2015 Min record')
plt.legend(prop={'size': 12})
fig.set_size_inches(18.5, 10.5)
plt.show()
```
