```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_html("https://en.wikipedia.org/wiki/History_of_the_Pakistani_cricket_team",skiprows=1,header=0)
%matplotlib inline
```

```python
world_cup=df[0]
t20_cup=df[1]
champions_trophy=df[2]
asia_cup=df[3]
economic_indicator=pd.read_csv("Economic Indicators.csv",skiprows=4)
world_cup.drop([12,13],inplace=True)
world_cup["Year"]=np.arange(1975,2023,4)
world_cup
```

```python
world_cup[['W','GP']] = world_cup[['W','GP']].apply(pd.to_numeric, errors='coerce')
t20_cup[['W','GP']] = t20_cup[['W','GP']].apply(pd.to_numeric, errors='coerce')
champions_trophy[['W','GP']] = champions_trophy[['W','GP']].apply(pd.to_numeric, errors='coerce')
asia_cup[['W','GP']] = asia_cup[['W','GP']].apply(pd.to_numeric, errors='coerce')
```

```python
t20_cup
t20_cup.drop([7,6],inplace=True)
t20_cup["Year"]=[2007,2009,2010,2012,2014,2016]
t20_cup
```

```python
GDP=economic_indicator[(economic_indicator["Indicator Name"]=="GNI (current US$)")]
government=economic_indicator[(economic_indicator["Indicator Name"]=="General government final consumption expenditure (% of GDP)")]
inflation=economic_indicator[(economic_indicator["Indicator Name"]=="Inflation, consumer prices (annual %)")]
GDP.drop(["Country Name",'Country Code','Indicator Name','Indicator Code'],1,inplace=True)
GDP=np.array(GDP.values.tolist()).reshape(-1,)
government.drop(["Country Name",'Country Code','Indicator Name','Indicator Code'],1,inplace=True)
government=np.array(government.values.tolist()).reshape(-1,)
inflation.drop(["Country Name",'Country Code','Indicator Name','Indicator Code'],1,inplace=True)
inflation=np.array(inflation.values.tolist()).reshape(-1,)
percentage_gdp=[]
a=-1
for i in GDP:
    if a!=-1:
        p=(i-prev_value)/prev_value
        percentage_gdp.append(p*100)
    prev_value=i
    a+=1
percentage_gdp
```

```python
world_cup["win_ratio"]=(world_cup["W"]/world_cup["GP"])*100
t20_cup["win_ratio"]=(t20_cup["W"]/t20_cup["GP"])*100
champions_trophy["win_ratio"]=(champions_trophy["W"]/champions_trophy["GP"])*100
asia_cup["win_ratio"]=(asia_cup["W"]/asia_cup["GP"])*100
```

```python
final_df=pd.DataFrame()
final_df["Years"]=np.arange(1983,2018,1)
```

```python
df=pd.concat([world_cup,t20_cup])
df=pd.concat([df,champions_trophy])
df=pd.concat([df,asia_cup])
df.reset_index(inplace=True)
```

```python
df.drop('index',1,inplace=True)
df
```

```python
df.drop([41,30,11],inplace=True)
df.drop(26,inplace=True)
df
```

```python
df["Year"] = df["Year"].apply(pd.to_numeric, errors='coerce')
df["GP"] = df["GP"].apply(pd.to_numeric, errors='coerce')
df["L"] = df["L"].apply(pd.to_numeric, errors='coerce')
df["T"] = df["T"].apply(pd.to_numeric, errors='coerce')
df["NR"] = df["NR"].apply(pd.to_numeric, errors='coerce')
df.dtypes
```

```python
df.sort_values(by=["Year"],inplace=True)
df.head()
```

```python
fig=plt.figure(figsize=(11,11))
plt.style.use('seaborn-notebook')
ax=fig.add_subplot(221)
ax.plot(df["Year"],df["win_ratio"],color='green', linestyle='dashed',linewidth=2, markersize=12,label="winning matches %")
ax.set_xlabel("Years")
ax.set_ylabel("Winning Percentage(%)")
ax.grid(False)
ax.set_facecolor("white")


ax1=fig.add_subplot(222)
ax.set_title("                           Comparsion of Pakistan's Cricket Team Performance")
ax1.plot(np.arange(1960,2018,1),percentage_gdp,color="red",linestyle='dashed',linewidth=2,markersize=12,label='GDP Growth')
ax1.set_xlabel("Years")
ax1.set_ylabel("GDP Growth(%)")
ax1.grid(False)
ax1.set_facecolor("white")



ax2=fig.add_subplot(223)
ax1.set_title("with Pakistan's Economy Performance")
ax2.plot(np.arange(1959,2018,1),inflation,color="blue",linestyle='dashed',linewidth=2,markersize=12,label='Inflation')
ax2.set_xlabel("Years")
ax2.set_ylabel("inflation-CPI(%)")
ax2.grid(False)
ax2.set_facecolor("white")


ax3=fig.add_subplot(224)
ax3.plot(np.arange(1959,2018,1),inflation,color="blue",linestyle='dashed',linewidth=2,markersize=12,label='Inflation')
ax3.plot(df["Year"],df["win_ratio"],color='green', linestyle='dashed',linewidth=2, markersize=12,label="winning matches %")
ax3.plot(np.arange(1960,2018,1),percentage_gdp,color="red",linestyle='dashed',linewidth=2,markersize=12,label='GDP Growth')
ax3.set_xlabel("Years")
ax3.set_ylabel("Percentages(%)-Each Indicator")
ax3.grid(False)
ax3.set_facecolor("white")
ax3.legend(facecolor="white")


plt.show()
```
